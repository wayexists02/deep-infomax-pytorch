```python
import numpy as np
import matplotlib.pyplot as plt

from utils import STLDataLoader
```

```python
train_loader = STLDataLoader(16)
```

```python
x = next(train_loader)
print(x.shape)
```

```python
x = np.transpose(x, (0, 2, 3, 1))
x = x*256 + 128
x = x.astype(np.uint8)
print(x.shape, x.max(), x.min())
```

```python
fig, axes = plt.subplots(4, 4, figsize=(12, 12))

for h in range(axes.shape[0]):
    for w in range(axes.shape[1]):
        i = h*axes.shape[1] + w
        axes[h, w].imshow(x[i])
        axes[h, w].set_xticks([])
        axes[h, w].set_yticks([])
        
plt.show()
```
